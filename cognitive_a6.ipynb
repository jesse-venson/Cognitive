```python
import numpy as np
import matplotlib.pyplot as plt

m=int(input('enter value of M:'))
x=np.linspace(-10,10)
y1=m*x**2
y2=m*np.sin(x)

plt.plot(x,y1,'b-',label='y=m*x^2')
plt.plot(x,y2,'r--',label='y=m*sin(x)')
plt.title('a6_q1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset of five subjects and scores
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Score': [85, 90, 78, 88, 95]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Set plot style
sns.set(style="whitegrid")

# Create the bar plot with different colors
colors = sns.color_palette("husl", len(df)) # for different colors for bars
barplot = sns.barplot(x='Subject', y='Score', data=df, palette=colors)

# Add annotations on top of each bar
for index, row in df.iterrows():
    barplot.text(index, row['Score'] + 1, row['Score'], color='black', ha="center")

# Add title, labels, and grid
plt.title("Scores in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Replace with your actual roll number
roll_number = 123456
np.random.seed(roll_number)

# Generate dataset
data = np.random.randn(50)

# Set up 2x2 subplot
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Cumulative sum (line plot)
axs[0, 0].plot(np.cumsum(data), color='blue')
axs[0, 0].set_title("Cumulative Sum Line Plot")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Cumulative Sum")
axs[0, 0].grid(True)

# Plot 2: Scatter with noise
noise = np.random.randn(50)
axs[0, 1].scatter(data, noise, color='green')
axs[0, 1].set_title("Scatter Plot with Random Noise")
axs[0, 1].set_xlabel("Data")
axs[0, 1].set_ylabel("Noise")
axs[0, 1].grid(True)

# Leave the remaining 2 subplots empty or turn them off
axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```
